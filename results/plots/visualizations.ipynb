```python
import matplotlib.pyplot as plt
from csv import reader
```

```python
import seaborn as sns
sns.set(color_codes=True)
```

```python
#Distribution of Top 20 complaints
x = []; y = []
with open('../res_use_cases/overall_complaint_type_dist.out','r') as f:
    rows = reader(f, delimiter = '\t')
    for i, r in enumerate(rows):
        x.append(r[0].title())
        y.append(int(r[1]))
        if i >= 19:
            break
            
plt.xticks(rotation=90)
g = sns.barplot(x,y)
g.set_title("Top 20 Complaints")
g.set_xlabel("Complaints ----->")
g.set_ylabel("Count ----->")
```

```python
#Distribution of number of complaints across the years
x = []; y = []
with open('../res_use_cases/year_dist.out','r') as f:
    rows = reader(f, delimiter = '\t')
    for i, r in enumerate(rows):
        x.append(r[0].title())
        y.append(int(r[1]))
        
g = sns.pointplot(x,y)
g.set_title("Count of Complaints across the Years")
g.set_xlabel("Year ----->")
g.set_ylabel("Count of Complaints ----->")
```

```python
#Distribution of types of complaints across the years
x = []; y = []; h = []
with open('../res_use_cases/year_complaint_type_dist.out','r') as f:
    rows = reader(f, delimiter = '\t')
    for i, r in enumerate(rows):
        x.append(r[0].split(',')[0].strip().title())
        h.append(r[0].split(',')[1].strip().title())
        y.append(int(r[1]))
        if i >= 39:
            break
            
g = sns.pointplot(x,y,hue=h)
g.set_title("Distribution of Top Complaint Types across the Years")
g.set_xlabel("Year ----->")
g.set_ylabel("Count of Complaints ----->")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
#Distribution of complaint status
x = []; y = []
with open('../res_use_cases/status_dist.out','r') as f:
    rows = reader(f, delimiter = '\t')
    for i, r in enumerate(rows):
        x.append(r[0].title())
        y.append(int(r[1]))
        
plt.xticks(rotation=90)
g = sns.barplot(x, y, estimator=lambda x: x*100.0/sum(y))
g.set_title("Distribution of Complaints Status")
g.set_xlabel("Status ----->")
g.set_ylabel("Percent of Complaints ----->")
```

```python
#Distribution of days to close a complaint by agency
x = []; y = []
with open('../res_use_cases/agency_closing_time_dist.out','r') as f:
    rows = reader(f, delimiter = '\t')
    for i, r in enumerate(rows):
        x.append(r[0])
        y.append(float(r[1]))
        if i >= 20:
            break
        

plt.xticks(rotation=90)
g = sns.barplot(x, y)
g.set_title("Distribution of Days to Close a Complaint By Agency")
g.set_xlabel("Agency ----->")
g.set_ylabel("Mean of Days to Close a Complaint ----->")
```
